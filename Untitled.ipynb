# **<span style="color:green"> Étape 4 : Entraînement des Modèles</span>** 

```python
# Choix des Modèles

models = [
    ('Logistic Regression', LogisticRegression()),
    ('Random Forest', RandomForestClassifier()),
    ('SVM', SVC(probability=True))  # Activer probability=True pour AUC-ROC
]
```

```python
# Choix des Modèles
models = [
    ('Logistic Regression', LogisticRegression()),
    ('Random Forest', RandomForestClassifier()),
    ('SVM', SVC(probability=True))  # Activer probability=True pour AUC-ROC
]

# Évaluation des Modèles
results = []
for name, model in models:
    # Validation croisée
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    # Entraînement du modèle
    model.fit(X_train, y_train)
    # Prédiction sur l'ensemble de test
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # Probabilité pour la classe positive

    # Métriques d'évaluation
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)

    # Courbe ROC
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.2f})')

    # Matrice de confusion
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))  # Ajuster la taille de la figure si nécessaire
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Matrice de Confusion pour {name}")
    plt.xlabel("Prédiction")
    plt.ylabel("Valeur Réelle")
    plt.show()

    results.append((name, accuracy, precision, recall, f1, auc, cv_scores.mean()))

```

```python
# Affichage des résultats
# Affichage des résultats
for name, accuracy, precision, recall, f1, auc, cv_mean in results:
    print(f'{name}:')
    print(f'  Accuracy: {accuracy:.2f}')
    print(f'  Precision: {precision:.2f}')
    print(f'  Recall: {recall:.2f}')
    print(f'  F1-score: {f1:.2f}')
    print(f'  AUC-ROC: {auc:.2f}')
    print(f'  Validation Croisée Moyenne: {cv_mean:.2f}')

    print('---------------------')
```

```python
# Sélection du Meilleur Modèle
best_model_name = results[np.argmax([result[5] for result in results])][0]  # Choisir le modèle avec la meilleure AUC
print(f'Meilleur modèle : {best_model_name}')
```
